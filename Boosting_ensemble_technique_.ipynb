```python
# Importing necessary libraries
import pandas as pd  # For data manipulation
from sklearn.datasets import load_iris  # For loading the Iris dataset
from sklearn.metrics import accuracy_score  # For model evaluation using accuracy score

# Loading the Iris dataset for AdaBoost implementation
dataset = load_iris()
X = dataset.data   # Feature variables
y = dataset.target # Target variable

# Displaying the shape of features and target
print(X.shape)  # Shape of X (features)
print(y.shape)  # Shape of y (target)
```

```python
# splitting the dataset into training and test sets for model evaluation
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Ada Boost classifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators=50)# making a model
```

```python
model.get_params()
```

```python
# fitiing the Adaboost model on the training data
model.fit(X_train, y_train)
```

```python
# perdiciting on teh test set
y_pred = model.predict(X_test)
```

```python
#evaluating model accuracy
print("AdaBoost Model Accuracy:",accuracy_score(y_test,y_pred))
```

```python
# gradient boosting implementation
from sklearn.ensemble import GradientBoostingClassifier
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
print(test_data.head())
```

```python
# dropping irrelevant columns for Gradient boosting
drop_columns = ['Name', 'Ticket', 'Cabin', 'Embarked','Parch','Age']
train_data.drop(labels=drop_columns, axis=1, inplace=True)
test_data.drop(labels=drop_columns, axis=1, inplace=True)
```

```python
# Encoding categorical data(sex column)
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
train_data['Sex'] = lb.fit_transform(train_data['Sex'])
test_data['Sex'] = lb.fit_transform(test_data['Sex'])
```

```python
# separating the target variable "survived" and feature variables
ytrain = train_data['Survived']
train_data.drop(labels='Survived', axis=1, inplace=True)
Xtrain = train_data
```

```python
from sklearn.model_selection import train_test_split
state = 12
X_train, X_test, y_train, y_test = train_test_split(Xtrain, ytrain, test_size=0.3,random_state=state)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lr_list = [0,.05,0.75,0.1,0.25,0.5,0.75,1]
for learning_rate in lr_list:
    gb_clf = GradientBoostingClassifier(n_estimators=20, learning_rate=learning_rate, max_features=2, max_depth=2, random_state=0)
    gb_clf.fit(X_train, y_train)

    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
```

```python
# final gradient boosting model with a chosen learning rate e
gb_clf2 = GradientBoostingClassifier(n_estimators=20, learning_rate=0.5, max_features=2, max_depth=2, random_state=0)
gb_clf2.fit(X_train, y_train)
```

```python
# predicting on the validation set
predictions = gb_clf2.predict(X_test)
```

```python
# evlauating model performance using confusion matrix and classification report
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# displaying the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test,predictions))
```
